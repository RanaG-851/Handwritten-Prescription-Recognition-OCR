```python
image = cv2.imread("prescription.jpeg")

# Define the crop area (x1, y1, x2, y2)
area = [100, 100, 200, 200]  # Example values

x = 0
y = 0
for i in range(100):
    img = image[area[1]:area[3]+x, area[0]:area[2]+y]

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f"Cropped Image {i+1}")
    plt.axis('off')
    plt.show()
    
    x = 21
    y = 22
```
